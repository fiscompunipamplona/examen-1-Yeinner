# 2.1

$(n,k) = \frac{n!}{k!(n-k)!}$

```python
from math import *
def binomial(n,k):
    
    if k==0:
        resultado=1
    elif k>=1:
        resultado=(factorial(n))/(factorial(k)*factorial(n-k))
    return (resultado)
n=int(input("Ingrese el valor de n: "))
k=int(input("Ingrese el valor de k: "))
b=binomial(n,k)
print(int(b))
```

# 2.2.a

```python
from math import *
def binomial(n,k):
    
    if k==0:
        resultado=1
    elif k>=1:
        resultado=(factorial(n))/(factorial(k)*factorial(n-k))
    return (resultado)
n=100
k=60
b=binomial(n,k)
P=b/(2**n)
print("La probabilidad de que caiga cara 60 veces es: ", P)
```

# 2.2.b

```python
from math import *
def binomial(n,k):
    if k<60:
        print("El numero es menor que 60")
        k=int(input("Ingrese un numero mayor que 60: "))
        resultado=(factorial(n))/(factorial(k)*factorial(n-k))
    elif k>=60:
        resultado=(factorial(n))/(factorial(k)*factorial(n-k))
        
    return (resultado)
n=100
k=int(input("Ingrese un numero mayor que 60: "))
b=binomial(n,k)
P=b/(2**n)
print("La probabilida de que caiga 60 o mas veces la misma cara es: ", P)
```
